```python
rows = 5

for i in range(1,rows+1):
    print('* ' * i)
    # print(i)
```

```python
for i in range (rows, 0 , -1):
    print('* ' * i)
    # print(i)
```

```python
for i in range(rows):
    print('* ' * rows)
```

```python
for i in range(1, rows+1):
    print(" " * (rows - i) + "*" * (2 * i - 1))
```

```python
for i in range(rows, 0, -1):
    print(" " * (rows - i) + "*" * (2 * i - 1))
```
