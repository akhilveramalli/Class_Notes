```python
# Generator

def gencube(n):
    for num in range(n):
        yield num ** 3

gencube(5)
```

```python
for i in gencube(9):
    print(i)
    
z = gencube(9)
next(z)
next(z)
next(z)

n = list(gencube(5))
n
```

```python
# Classes : Blueprints of objects 
# Class Methods : Functions inside the class are called as methods
# __init__ : Used to define variables inside the 

class car:
    def __init__(self, company, model, color): 
        self.company = company
        self.model = model
        self.color = color
    def get_company(self):
        return self.company
    def __str__(self):
        return (f"Car Company: {self.company} , Model: {self.model}, Color: {self.color} ")


x = car("Honda", "Civic","White")
print(x)
print(x.get_company())

print(x.company)
        
```

```python
# Abstraction : Protecting the class variables from direct access
class car:
    def __init__(self, company, model, color): 
        self.__company = company
        self.__model = model
        self.__color = color
    def get_company(self):
        return self.__company
    def __str__(self):
        return (f"Car Company: {self.__company} , Model: {self.__model}, Color: {self.__color} ")


x = car("Honda", "Civic","White")
print(x)
print(x.get_company())

print(x.__company)
```

```python
# Inheritance: Inheritance allows one class to reuse properties of another class.

# Child class can use the variables and methods of parent class

class animal: # Parent Class
    def speak(self):
        print("Animal makes sound")

class dog(animal): # child class(dog) inheriting parent class(animal) 
    def bark(self):
        print("Dog barks")

d = dog()

d.bark()
d.speak()
```
